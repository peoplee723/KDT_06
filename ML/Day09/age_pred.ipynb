```python
def draw_bar(df):
    col = 3
    row = (len(df.columns) + col - 1) // col
    colors = [
    "#AED6F1",  # 옅은 하늘색
    "#A9DFBF",  # 옅은 초록색
    "#F9E79F",  # 옅은 노란색
    "#F5B7B1",  # 옅은 핑크색
    "#D2B4DE",  # 옅은 보라색
    "#FAD7A0",  # 옅은 주황색
    "#D5DBDB",  # 옅은 회색
    "#A3E4D7",  # 옅은 민트색
    "#D7BDE2",  # 옅은 연보라색
    "#F5CBA7",  # 옅은 살구색
]

    plt.figure(figsize=(15, row * 4))
    for i, column in enumerate(df.columns, 1):
        plt.subplot(row, col, i)
        df[column].value_counts().plot(kind='bar', color=colors)
        plt.title(column)
        plt.xticks(rotation=360)
    plt.tight_layout()
    plt.show()
```

```python
# study hour-> nan값은 0으로, -값은 제거
dataDF['Study Hours per Week'].fillna('0')
search= dataDF['Study Hours per Week']<0
num= dataDF[search].index
dataDF.drop(index=num, inplace=True)
dataDF['Study Hours per Week'].describe()
```

```python
# attendance_rate-> nan, -값, 100초과값 모두 제거
dataDF['Attendance Rate'].dropna(inplace=True)
search1= dataDF['Attendance Rate']<0 
search2= dataDF['Attendance Rate']>100
num=dataDF[search1].index
dataDF.drop(index=num, inplace=True)
dataDF=dataDF.reset_index(drop=True)
num2=dataDF[search2].index
dataDF.drop(index=num2, inplace=True)
dataDF['Attendance Rate'].describe()
dataDF=dataDF.reset_index(drop=True)
dataDF['Attendance Rate'].describe()
```
