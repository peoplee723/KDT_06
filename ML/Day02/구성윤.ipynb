```python
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 기본 -> ndarray , as_frame=True-> 데이터프레임 형태로
dataDict= load_diabetes(as_frame=True)
```

```python
# 기본 -> ndarray , return_X_y=True-> tuple로 data와 target만 반환
# dataDict= load_diabetes(return_X_y=True)
```

```python
print(dataDict.keys())
# print('DESCR')
```

```python
# 데이터 불러오기
DATA_FILE= '../data/diabetes.csv'
data_DF=pd.read_csv(DATA_FILE)
```

```python
# 상관관계 확인
data_DF.corr()
# BP, ST, Insulin, Glucose
```

```python
DF= dataDict['data']
DF['glucose']=dataDict['target']
print(DF.head())
```

```python
# 상관관계 확인
DF.corr()
# bmi,s5, bp
```

```python
# 데이터 분리
features= DF[['bmi', 's5', 'bp']]
target=DF['glucose']
print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, Y_train, Y_test= train_test_split(features, target, test_size=.3,
                                                    random_state=10)
```

```python
# 모델 인스턴스 생성
from sklearn.neighbors import KNeighborsRegressor
model= KNeighborsRegressor()
# 학습
model.fit(X_train, Y_train)
```

```python
# 최적의 K 찾기
optima={}
for i in range(1, model.n_samples_fit_+1):
    model.n_neighbors=i

    score= model.score(X_test, Y_test)
    optima[i]=score

sorted(optima.items(), key=lambda x: x[1], reverse=True)
```

```python
# 학습 K=21
model.n_neighbors=21
model.fit(X_train, Y_train)
```

```python
# 평가
model.score(X_test, Y_test)
```
