```python
# 폴더 안에 파일 및 폴더를 가져오기
def make_dataset(dir_path, file_path):
    files=os.listdir(dir_path)
    dataDF=pd.DataFrame()
    for file in files:
        print(file, dir_path+file, os.path.exists(dir_path+file))
        label=file[:2]  #(en,fr... 추출)
        print(label)
        with open(dir_path+file, mode='r') as f:
            all_data=f.read()
            all_data=all_data.lower()       #대소문자 통일
            all_data=all_data.replace('\n', '')  #줄바꿈 제거

            for ch in all_data:
                if (ord('a')> ord(ch)) or (ord(ch) >ord('z')):      #알파벳 제외하고 전부 제거
                    all_data=all_data.replace(ch, '')
            
        # a-z 개수 카운팅
        num=dict(Counter(all_data))
        with open(file_path, mode='a') as f:
            cc=dict(Counter(all_data))
            f.write(label+',')
            f.write(','.join([str(value) for value in cc.values()]))
        print(all_data)
        print(len(all_data))
        print(num)
        print(dataDF)
    
    
    
```

```python
# 폴더 안에 파일 및 폴더를 가져오기-> 데이터 프레임으로 변환후 저장
files=os.listdir(TRAIN_PATH)
dataDF=pd.DataFrame()
for file in files:
    print(file, TRAIN_PATH+file, os.path.exists(TRAIN_PATH+file))
    label=file[:2]  #(en,fr... 추출)
    print(label)
    with open(TRAIN_PATH+file, mode='r') as f:
        all_data=f.read()
        all_data=all_data.lower()       #대소문자 통일
        all_data=all_data.replace('\n', '')  #줄바꿈 제거

        for ch in all_data:
            if (ord('a')> ord(ch)) or (ord(ch) >ord('z')):      #알파벳 제외하고 전부 제거
                all_data=all_data.replace(ch, '')
        
    # a-z 개수 카운팅
    num=dict(Counter(all_data))
    num_df=pd.DataFrame(data=num.values(), columns=num.keys(), index=[file])        #데이터 프레임으로 변환
    num_df=num_df.sort_index(ascending=True)            #정렬-> items로 튜플로 꺼내서 변환 or 데이터프레임으로 만든 뒤에 정렬
    num_df=num_df/len(all_data)             #비율로 계산
    dataDF[file]=num_df
    # with open('backup.csv', mode='w') as f:
    #     f.write(num)
    # dataDF[file]= num

    print(all_data)
    print(len(all_data))
    print(num)
    print(dataDF)
    dataDF.to_csv('lan_Data_train.csv')
    
    
```
