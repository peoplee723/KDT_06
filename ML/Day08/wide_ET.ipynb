#### Ensemble - RandomForest & ExtraTree
- 배깅 방식의 앙상블: 중복 허용한 랜덤 샘플 + 동일 모델
    - ex) RandomForestC/R
- 페이스트 방식의 앙상블: 중복 허용 안한 랜덤 샘플 + 동일 모델
    - ex) ExtraTreeC/R

```python
# [1] 데이터 준비
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 데이터
Data= '../Data/wine.csv'
WineDF= pd.read_csv(Data)
WineDF.info()
```

```python
WineDF.head()
```

```python
WineDF.describe() #랜덤 포레스트라 스케일링은 선택
```

```python
# [2] 학습준비
# 데이터셋 분할
feature=WineDF[WineDF.columns[:-1]]
target=WineDF[WineDF.columns[-1]]


print(f'{feature.shape}, {target.shape}')

X_train, X_test, Y_train, Y_test= train_test_split(feature,target, stratify=target, random_state=1,
                                                   test_size=.2)
```

```python
# 인스턴스 생성: 100개의 내부 DT모델에서 사용할 데이터셋 생성
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.
lf_model= ExtraTreesClassifier(random_state=10, ) ##oob는 부트스트랩 True일 경우만 사용가능!
lf_model.fit(X_train,Y_train)
```

```python
# 모델 파라미터
print(lf_model.classes_, lf_model.n_classes_, lf_model.feature_names_in_, 
      lf_model.n_features_in_, lf_model.feature_importances_, 
      
      sep='\n')
# bootstrap, oob_score, Max_sample
```

```python
print(lf_model.estimator_)
for est in lf_model.estimators_:
    print(est)
```

```python
train_score= lf_model.score(X_train, Y_train)
test_score= lf_model.score(X_test, Y_test)
print( train_score, test_score)
```

```python
# [5] 튜닝 
from sklearn.model_selection import RandomizedSearchCV
#지정된 범위에서 지정된 횟수만큼 하이퍼파라미터를 추출+ 조합하여 진행
# ==> 하이퍼 파라미터 정확이 얼마 줘야하는지 모를 때 사용
```

```python
# 하이퍼 파라미터 설정 (꼭 range형태가 아니여도 됨!)
params={'max_depth': range(2,30),
        'min_samples_leaf': range(5,30)
        }
```

```python
rf_model= ExtraTreesClassifier(random_state=7, n_estimators=300)
searchCV= RandomizedSearchCV(rf_model, param_distributions= params, verbose=3, n_iter=100)
searchCV.fit(X_train, Y_train)
```

```python
# 모델 파라미터
print(searchCV.best_score_, searchCV.best_params_, searchCV.best_estimator_, sep='\n')
df=pd.DataFrame(searchCV.cv_results_)
```
