```python
import pandas as pd
import numpy as np
```

```python
file_name=(r'C:\Users\KDP-25\Desktop\KDT_06\numpy\Day01\과제1-데이터예시(seoul.csv).csv')
df= pd.read_csv(file_name)
df
```

```python
# 1. 각 지하철 역별 평균 이용객수
# 역명 추출
station_name=set(df['역명'])
print(station_name)
station_dict=dict()
# dict로 저장
for station in station_name:
    station_dict[station]=0
print(station_dict)



```

```python
# 평균이용객수 계산(총 승하차이용객수)
station_total_dict= station_dict.copy()
for i in df.index:
    name= df.iloc[i]['역명']
    current=station_dict[name]
    station_total_dict[name]= (current+df.iloc[i]['승하차객_2'])
# print(station_dict)
# 평균
station_average_dict=station_total_dict.copy()
for station in station_average_dict:
    station_average_dict[station]= station_average_dict[station]/31
print(f'{station_average_dict}')
```

```python
search= df['역명']=='강일'
search_list=df[search]['승하차객수'].std(axis=0)
print(search_list)
```

```python
# 2. 역별 표준편차
for station in station_dict:
    search= df['역명']==station
    # print(df[search]['승하차객수'].std())
    station_dict[station]=(df[search]['승하차객_2'].std())
search_dict=([station_dict])
print(search_dict)
```
