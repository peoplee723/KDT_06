#### [파이 그래프]
- 두개의 컬럼(변수)의 관계를 시각화해서 표현
-연속형(수치형) 데이터에 사용함

```python
# [1]
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2]
file_name= r'C:\Users\KDP-25\Desktop\EX_PANDAS06\data\auto_mpg.csv'
```

```python
# [3]
dataDF= pd.read_csv(file_name)
```

```python
# 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# 자동차 생산 국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# 자동차 생산국에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
# 자동차 생산국가에 대한 비율 시각화-> bar 그래프
x=dataDF['origin'].unique() 
h=dataDF['origin'].value_counts()


origin_bar= plt.bar(x, h, color= ['r', 'b', 'g'], edgecolor= 'k')
plt.xticks([1,2,3], ['US', 'EU', 'JP'], rotation= 90, fontsize= 20)
plt.bar_label(origin_bar, h, label_type='edge')
```

```python
# 파이 그래프

plt.pie(h, labels=['US', 'EU', 'JP'], autopct='%.1f%%')
plt.show()
```

```python
# 파이 그래프

plt.pie(h, labels=['US', 'EU', 'JP'], autopct='%.1f%%', counterclock=False,
        startangle=300)
plt.show()
```

```python
plt.pie(h, labels=['US', 'EU', 'JP'], autopct='%.1f%%', counterclock=False,
        startangle=300, explode=[0.1, 0, 0])
plt.show()
```

```python
plt.pie(h, labels=['US', 'EU', 'JP'], autopct='%.1f%%', counterclock=False,
        startangle=300, wedgeprops={'linewidth':10, 'width':.7, 'edgecolor': 'w'},
        shadow=True)
plt.show()
```
