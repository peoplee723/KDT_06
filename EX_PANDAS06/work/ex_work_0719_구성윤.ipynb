```python
df= pd.read_csv(r'C:\Users\people\Desktop\KDT\coding\work\bookings\bookings.csv')

#결측치 제거 (기본=행)
df= df.dropna()
df.info()
```

```python
df['Total_Review']= df['Total_Review'].map(lambda x: str(x).replace('external','').strip())   #열 내 데이터를 하나씩 꺼내서 문자 및 공백 제거
df['Total_Review']= df['Total_Review'].map(lambda x: str(x).replace('review','').strip())
df['Total_Review']= df['Total_Review'].map(lambda x: str(x).replace(',',''))
df['Total_Review']= df['Total_Review'].astype('float64')

```

```python
# 분위수 구하기 


quantile= [0, 0.2, 0.4, 0.6, 0.8, 1]

for idx in quantile:
    q= df['Total_Review'].quantile(idx, interpolation='lower') #interpolation=linear(기본)
    print(f'quantile({idx}) is {q}')

# 0.2단위로 분위를 나눠서 분위수 도출

for idx in quantile:
    q= df['Total_Review'].quantile(idx, interpolation='nearest') #interpolation=linear(기본)
    print(f'quantile({idx}) is {q}')
```

```python
import seaborn as sns
df= sns.load_dataset('penguins')
df.head()
```

```python
#수치형 변수 필터링

df.query('bill_length_mm> 55')
```

```python
# boolean indexing으로도 필터링 가능 ---> query가 좀 더 짧음

df.loc[df['bill_length_mm']> 55]
```

```python
# and or not 연산자도 사용가능

df.query('bill_length_mm> 55 and species== "Gentoo"')
```

```python
# 외부 변수 참조시 @변수명 사용

length= 55; species= 'Gentoo'

df.query('bill_length_mm>= @length and species == @species')
```

```python
# 문자열 필터링 시 -->contains(), startswith, endswith 메서드 사용

df.query('island.str.contains("oe")', engine='python')   #TypeError 발생시 엔진=파이썬 추가
```

```python
df.query('species.str.endswith("e")')
```

```python
filtering= ['Adelie', 'chinstrap']
df.query('species.isin(@filtering)')
```
