```python
import pandas as pd
file_name='titanic.csv'
df=pd.read_csv(file_name)
```

```python
df.describe()
# 평균 지불 금액=32.2
```

```python
df.isna().sum()
```

```python
# 금액을 기준으로
# 1. 생존자가 지불한 평균 비용  --->생존자=342명, ,평균 지불 비용 48.4
index_alive= df['alive']=='yes'
df[index_alive]['fare'].mean()
```

```python
# 생존자 성비
df[index_alive]['sex'].value_counts()
# 남=109, 여=233                               (총인원-> 남=577, 여=314  --->
                                                #  남=537, 여=271, 아이=83->)
# 남 생존율=19%, 여 생존율=74%                      #남아=40, 여아=43
# 아이 83명 중 49명 생존(생존율=59%)


# 성인 남=88 성인여=205
# 성인 남 생존율=16%, 성인 여 생존율=76%
df[index_alive]['who'].value_counts()
```

```python
index_child_male= (df['who']== 'child') & (df['sex']=='male') 
df[index_child_male]['survived'].value_counts()
# 남아 40중 21명 생존-> 생존율= 53%
```

```python
index_child_female= (df['who']== 'child') & (df['sex']=='female')
df[index_child_female]['survived'].value_counts()
# 여아 43명중 15명 생존 ->생존율=65%
```

```python
# 생존자와 사망자 간의 지불 금액의 차이가 있는가?
# 1. 전체 인원의 비용 통계
df['fare'].describe()
# 평균 32.2   최대=512.32, 최소=0
```

```python
# 생존자 통계
df[index_alive]['age'].describe()
# 나이-> 평균 28세, 최소 0.4세, 최대 80세
# 생존자 구성->성인 남=88명, 성인여=205명, 아이=49명(남아=21, 여아=28)
```

```python
# 2. 생존자의 비용 통계
df[index_alive]['fare'].describe()
# 생존자 평균 비용=48.4, 최대=512, 최소=0
```

```python
# 3. 사망자의 비용 통계
df[~index_alive]['fare'].describe()
# 사망자 평균 비용=22.11, 최대=263 최소=0

# 즉 생존자가 평균적으로 더 많은 비용을 지불하였다.
```

```python
# 최적의 생존비용
# 1. 지불한 금액을 3단위로 분리
mpower= pd.cut(df['fare'], bins=3, labels=['저','중','고'])
# 2. 컬럼에 추가
df['money_power']=mpower

```

```python
# 분류별 생존율
index_money= df['money_power']=='저'
df[index_money]['alive'].value_counts()
# 저=> 328/871  =>38%
328/871
df[index_money].describe()
```

```python
# 중
index_money= df['money_power']=='중'
df[index_money]['alive'].value_counts()
# 중-> 11/17  ->65%
11/17
df[index_money].describe()
```

```python
# 고
index_money= df['money_power']=='고'
df[index_money]['alive'].value_counts()
# 고 --> 3/3  ->100% , 1클래스, 남2,여1 도착지: Cherbourg 
df[index_money].describe()
```

```python
# 결론
# 많은 돈을 지불한 승객의 생환율이 높다
```
