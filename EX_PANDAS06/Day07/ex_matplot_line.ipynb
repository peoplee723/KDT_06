#### [Matplotlib 활용한 시각화]
-  Matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(line) 그래프
# plot(y축 데이터)
plt.plot([1,2,3,4])
plt.xlabel('Data')
plt.ylabel('Result')
plt.show()
```

```python
# [2] 그래픝 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터)
plt.plot([10,20,30], [5, 10, 15])
plt.show()
```

```python
# [2] 그래픝 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf-> 포맷 스트링) 
# fmf-> 그래프의 형태, 색상, 두께 등을 지정하는 스트링
plt.plot([10,20,30], [5, 10, 15],'b>--')
plt.show()
```

```python
# [2] 그래픝 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf-> 포맷 스트링) 
# fmf-> 그래프의 형태, 색상, 두께 등을 지정하는 스트링
# color, marker, linestyle, markersize, linewidth... 등
# markeredgecolor->mec 등 줄임말도 있음
# 필요할 때 찾아서 사용해야함 (외우기에는 너무 많음)

plt.plot([10,20,30], [5, 10, 15], 
         color= 'r', marker='o', linestyle= '--', 
         markersize=10, linewidth= 4, 
         mec='blue', mew= 5)
plt.show()
```

```python
# 그래프 그리기 ->선 그래프
plt.plot([10,20,30], [15,25,35], 'bo-')
plt.plot([10,20,30], [18,29,25], 'ro-')
plt.show()
```

```python
# 그래프 그리기 ->선 그래프
plt.plot([10,20,30], [15,25,35], 'bo-', 
         [10,20,30], [18,29,25], 'ro-')        #한 플롯안에 여러 선 그리기 가능
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# 그래프 제목, x/y축 라벨, 여러개 그래프 범례
# plot의 라벨 속성과 legend 함수가 연결되어 있음()
# -> 그래프의 label속성값을 범례에서 사용함

plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label='v')
plt.plot([10, 20, 30], [5, 30, 28], 'bo--', label='2')

plt.title('[MPLIB LINE2D]')
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# 그래프 제목, x/y축 라벨, 여러개 그래프 범례
# plot의 라벨 속성과 legend 함수가 연결되어 있음()
# -> 그래프의 label속성값을 범례에서 사용함
#       # loc 속성으로 범례 위치 지정 가능

plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label='v')
plt.plot([10, 20, 30], [5, 30, 28], 'bo--', label='2')

plt.title('[MPLIB LINE2D]')
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend(loc= 'upper center')
plt.show()
```
