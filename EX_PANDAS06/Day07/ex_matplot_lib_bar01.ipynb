#### [Matplotlib 활용 시각화] - 막대그래프
- 가장 많이 사용되는 그래프
- 데이터를 비교할 때 사용

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
plt.bar(fruits, counts, width=0.3, color= bar_colors, label= bar_labels)
plt.legend()
plt.show()
```

```python
# 수평 막대그래프          #차이--> height, width(너비 높이 차이)
plt.barh(fruits, counts, height=0.3, color= bar_colors, label= bar_labels)
plt.legend()
plt.show()
```

```python
plt.bar(fruits, counts, width=0.3, color= bar_colors, label= bar_labels)
plt.plot(counts, 'bo-')
plt.legend()
plt.show()
# 형태만 다르고 다른건 거의 동일함
```
