#### [Matplotlib 활용 데이터 시각화]
- x출과 y축 다양한 처리

```python
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기 ->축 범위 설정
y_data= [1,3,5,7,9]
x_data= ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot( x_data, y_data, marker='o')
plt.ylim(bottom= 0, top=15)
# plt.xlim( left=0, right=5)
plt.show()
```

```python
# 그래프 그리기  ->눈금 설정 
y_data= [1,3,5,7,9]
x_data= ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot( x_data, y_data, marker='o')
plt.ylim(bottom= 0, top=15)
# plt.xlim( left=0, right=5)
plt.xticks(rotation= 'vertical')  #vertical 또는 수치(각도)
plt.show()
```

```python
# 그래프 그리기  ->그래프 사이즈 설정
y_data= [1,3,5,7,9]
x_data= ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.figure(figsize=[8,8])
plt.plot( x_data, y_data, marker='o')
plt.ylim(bottom= 0, top=15)
plt.xticks(rotation= 'vertical')  #vertical 또는 수치(각도)
plt.show()
```

```python
# 그래프 그리기  ->사이즈 자동맞춤
y_data= [1,3,5,7,9]
x_data= ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot( x_data, y_data, marker='o')
plt.ylim(bottom= 0, top=15)
plt.xticks(rotation= 'vertical')  
plt.tight_layout()                  #크기 자동맞춤
plt.show()

```

```python
# 그래프 그리기  -> 눈금(ticks) 설
y_data= [1,3,5,7,9]
# x_data= ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.plot( y_data, marker='o')
plt.xlim(left= 0, right=7)
plt.xticks([0,2,4,6])
plt.tight_layout()                  #크기 자동맞춤
plt.show()

```
