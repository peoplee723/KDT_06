### DataFrame에서 행/열 추가

```python
import pandas as pd

data={'name': ['홍길동', '마징가', '원더우먼'],
      'job': ['의적', '로봇', '영웅'],
      'gender': ['남', '남', '여'],
      'age': [100,200,300], 
      'height': [180,300,200]}
 
```

```python
# DF으로 데이터 저장
dataDF=pd.DataFrame(data)
dataDF
```

-  실습 -> 행 추가: 변수명.loc['추가할 행이름']=데이터
    * 변수명.iloc[] 사용 불가
    * 데이터수==컬럼수 -> 행 추가 가능
    * 데이터수==1 -> 행 추가 가능
    * 그외 -> 행 추가 X

```python
# 1개 행 추가  => 데이터 수가 열 개수와 동일한 경우
dataDF.loc[3]=['베트맨', '박쥐', '남', 40, 190]        
dataDF
# iloc ->target object보다 크면 안됨 -->행 추가 수단으로 사용 불가능
```

```python
# 1개 행 추가  => 데이터 수가 한개 ->동일한 걸로 나옴
dataDF.loc[4]=100       
dataDF
```

```python
# dataDF.loc[6]=['스파이더맨','거미']  ->mismatch error-> 안됨    
dataDF
```

-  실습 -> 열 추가: 변수명['추가할 열이름'] = 데이터
    * 데이터수==행 수 -> 열 추가 가능
    * 데이터수==1 -> 열 추가 가능
    * 그외 -> 열 추가 X

```python
dataDF               
```

```python
# 컬럼/열 추가 ->weight 추가   (행 수== 데이터 수)
dataDF['weight']= [70, 3000, 60, 80, 100]
dataDF
```

```python
# 컬럼/열 추가 ->weight 추가   (데이터 수==1)
dataDF['weight2']= 0
dataDF
```
